```python
import pandas as pd

df = pd.read_excel('/content/Datathon Dataset (1).xlsx', dtype={'Offst.acct': str})
df.head()
```

```python
unique_vals = df['Offst.acct'].astype(str).str.strip().unique()
print(f"Total: {len(unique_vals)}")

targets = {'7177MY', '7327KR'}
found = targets.intersection(unique_vals)
missing = targets.difference(unique_vals)

print(f"Found: {found}")
print(f"Missing: {missing}")
print(unique_vals)
```

```python
clean_series = df['Offst.acct'].astype(str).str.strip()
mapping = df.groupby(clean_series)['Name of offsetting account'].unique()

for acct, names in mapping.items():
    print(f"Offst.acct: {acct}")
    print(*(f"  - {n}" for n in names), sep="\n")
    if len(names) > 1:
        print("  WARNING: Multiple names found!")
    print()

if (mapping.apply(len) == 1).all():
    print("All 'Offst.acct' values have unique names.")
```

```python
missing_rows = df[df['Category Index'].isna()]
print(f"Missing count: {len(missing_rows)}")
display(missing_rows)
```

```python
# 1. Define Static Mappings
category_declaration = {
    'Non Netting AP': 6,
    'Non Netting AR': 18,
}

# 2. Apply Static Map (preserves existing values if map returns NaN)
df['Category Index'] = df['Category'].map(category_declaration).fillna(df.get('Category Index', pd.Series([None]*len(df))))

# 3. Handle 'Other' Dynamic Logic (+ve -> 16, -ve -> 2)
mask_other = df['Category'] == 'Other'

# Apply overrides
df.loc[mask_other & (df['Amount in USD'] > 0), 'Category Index'] = 16
df.loc[mask_other & (df['Amount in USD'] < 0), 'Category Index'] = 2
# df.loc[mask_other & (df['Amount in USD'] == 0), 'Category Index'] = 22 # Optional fallback

# 4. Verify
print("Categories found:", df['Category'].dropna().unique())
print("Rows missing Category Index:", df['Category Index'].isna().sum())
display(df.head())
```

```python
df['Category Index'] = pd.to_numeric(df['Category Index'], errors='coerce')

df.loc[(df['Category Index'] >= 1.0) & (df['Category Index'] <= 15.0), 'Flow_Code'] = 1

df.loc[(df['Category Index'] >= 16.0) & (df['Category Index'] <= 25.0), 'Flow_Code'] = 2

print(df['Flow_Code'].value_counts(dropna=False))
```
